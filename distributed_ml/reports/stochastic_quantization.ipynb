```python
import matplotlib.pyplot as plt
```

```python
def parse_float(num):
    return float(num.strip())
```

```python
def parse_line(line):
    args = line.split(' ')
    return [args[0], list(map(parse_float, args[1:]))]
```

```python
def read_accs(file_path):
    with open(file_path, 'r') as f:
        lines = f.readlines()
        return list(map(parse_line, lines))
```

```python
def get_range(accs):
    n = len(accs)
    return list(range(n))
```

```python
def show_results(exps):
    plt.rcParams["figure.figsize"] = (20,10)
    for exp in exps:
        [name, accs] = exp
        plt.plot(get_range(accs), accs, label=name)
    plt.legend(loc=4, prop={'size': 15})
    plt.show()
```

```python
exps = read_accs('../logs/stochastic_quantization/stochastic_quantization.log')
show_results(exps)
```
